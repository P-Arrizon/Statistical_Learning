## The Model



Simple linear regression is a method for modeling the relationship between a continuous response variable and a single predictor variable. The goal is to fit a linear equation to the data that can be used to make predictions about the response variable based on the predictor variable.

The model assumes that the relationship between the predictor variable x and the response variable y can be represented by a straight line with an intercept (β₀) and slope (β₁):

\begin{equation}
y = \beta_0 + \beta_1 x + \epsilon
\end{equation}

where ε is the error term, which represents the variability in the response variable that is not explained by the predictor variable.



## Estimating the Variables

The goal of simple linear regression is to estimate the values of the intercept and slope that best fit the data. This is typically done using the method of least squares, which involves finding the values of β₀ and β₁ that minimize the sum of squared residuals:

\begin{equation}
\sum_{i=1}^n (y_i - \hat{y}i)^2 = \sum{i=1}^n (y_i - \beta_0 - \beta_1 x_i)^2
\end{equation}

where $ŷ_i$ is the predicted value of the response variable for the ith observation.

The least squares estimates of β₀ and β₁ can be calculated using the following formulas:

\begin{equation}
\hat{\beta}1 = \frac{\sum{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
\end{equation}

\begin{equation}
\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}
\end{equation}

where $\bar{x}$ and $\bar{y}$ are the sample means of the predictor variable and response variable, respectively.

Once the values of β₀ and β₁ have been estimated, they can be used to make predictions about the response variable for new values of the predictor variable.

Simple linear regression can also be used to test hypotheses about the relationship between the predictor variable and response variable. This typically involves calculating a t-statistic to test the null hypothesis that the slope of the regression line is zero:

\begin{equation}
t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}
\end{equation}

where SE(β₁) is the standard error of the slope estimate. If the t-statistic is large enough, the null hypothesis can be rejected in favor of the alternative hypothesis that the slope is not zero.

In summary, simple linear regression is a useful tool for modeling the relationship between a continuous response variable and a single predictor variable. The model can be used to make predictions about the response variable, as well as to test hypotheses about the relationship between the predictor variable and response variable.

```python
import numpy as np                                        
import scipy.stats as st                                    
import pandas as pd                                       
import matplotlib.pyplot as plt                        

df = pd.read_csv(r"https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/Density_Por_data.csv") # load data from Dr. Pyrcz's GitHub repository
df.describe().transpose()  

porosity = df['Porosity']
density = df['Density']
den = np.linspace(min(density),max(density), len(density)) # X domain

```

```python
slope, intercept, r_value, p_value, std_err = st.linregress(density,porosity)                     # output is the linear regression model

print('The model parameters are, slope (b1) = ' + str(round(slope,2)) + ', and the intercept (b0) = ' + str(round(intercept,2)))
plt.scatter(density, porosity, color = 'red',edgecolor='black',label='sample data')
plt.plot(den, intercept + slope*den, label='linear regression model')
plt.title('Sample Data and Model'); plt.xlabel('Density (g/cm3)'); plt.ylabel('Porosity (%)')
plt.legend(); plt.grid()
plt.show()

# Calculate the R-squared value
r_squared = r_value**2

# Calculate the t-value
t_value = slope / std_err

# Print the slope, intercept, R-squared and t-value
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-squared: ", r_squared)
print("t-value: ", t_value)
```
